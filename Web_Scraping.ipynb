```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue#2021_list'

page = requests.get(url)

soup = BeautifulSoup(page.text , 'html')
```

```python
table = soup.find_all('table')[1]
label = table.find_all('th')
title = [i.text.strip() for i in label]
df = pd.DataFrame(columns = title)
```

```python
label = table.find_all('th')
```

```python
df = pd.DataFrame(columns = title)
df
```

```python
data = table.find_all('tr')
[i.text.strip() for i in data]
```

```python
for row in data[1:]:
  row_data = row.find_all('td')
  values = [val.text.strip() for val in row_data]
  value = values[:1] + values[2:]
  #print(value)
  length = len(df)
  df.loc[length] = value
```

```python
#df['year'] = 2021
```

```python
df.rename(columns = {
    'Revenue ($B) USD[2]':'Revenue' ,
    "Employees[2]" : "Employees" ,
    "Revenue per employee ($K USD)[2]" : "Revenue per employee"
    }, inplace = True)
```

```python
df
```

```python
table2 = soup.find_all('table')[2]

label2 = table2.find_all('th')

title2 = [i.text.strip() for i in label2]

df2 = pd.DataFrame(columns = title2)

data = table2.find_all('tr')

for row in data[1:]:
  row_data = row.find_all('td')
  values = [val.text.strip() for val in row_data]
  value = values[:1] + values[2:]
  #print(value)
  length = len(df2)
  df2.loc[length] = value
```

```python
df2['year'] = 2020
```

```python
df2.rename(columns = {'Revenue ($B) USD[6]':'Revenue' , "Employees[6]" : "Employees" , "Revenue per employee ($K USD)[6]" : "Revenue per employee"}, inplace = True)
```

```python
table3 = soup.find_all('table')[3]

label3 = table3.find_all('th')

title3 = [i.text.strip() for i in label3]

df3 = pd.DataFrame(columns = title3)

data = table3.find_all('tr')

for row in data[1:]:
  row_data = row.find_all('td')
  values = [val.text.strip() for val in row_data]
  value = values[:1] + values[2:]
  #print(value)
  length = len(df3)
  df3.loc[length] = value
```

```python
df3['year'] = 2019
df3.rename(columns = {'Revenue ($B) USD[8]':'Revenue' , "Employees[8]" : "Employees" , "Revenue per employee ($K USD)[8]" : "Revenue per employee"}, inplace = True)
```

```python
combined_df = pd.concat([df,df2,df3])
combined_df
```

```python
combined_df['Revenue'] = combined_df['Revenue'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
combined_df['Revenue per employee'] =  combined_df['Revenue per employee'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
combined_df['year'] = pd.to_datetime(combined_df['year']).dt.year
```

```python
combined_df['Company'] = combined_df['Company'].replace({'Meta(Facebook)' : 'Meta'})
```

```python
combined_df
```

```python
data = pd.DataFrame(combined_df)
data
```

```python
data.to_excel('combined_table.xlsx' , index = False)
```
